# INTRODUCTION
Data visualization is the discipline of trying to understand data by placing it in a visual context so that patterns, trends, and correlations that might not otherwise be detected can be exposed.

I add weather informations, such as temperature and precipitations and Analyzed weather condition of different date/time at every 1 hours and compared them against different parameters like temp,feelslike,dew,humidity,precip,precipprob,preciptype,snow
sealevelpressure,cloudcover,visibility,solarradiation,solarenergy,uvindex,	severerisk,conditions,icon,stations.summarized all the results in one graphical plot.

I Used various Python libraries like Pandas,NumPY for organizing data and used Seaborn,Matplotlib and other relavant packages for visualization.

I have taken the dataset from https://www.weatherapi.com/


```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
weather = pd.read_csv('weather.csv')
```

```python
weather.head()
```

```python
weather.info()
```

```python
sns.barplot(x=weather['humidity'],y=weather['feelslike'])
```

```python
sns.displot(x=weather['humidity'])
```

```python
sns.displot(x=weather['humidity'], kde=False, rug=True);
```

```python
sns.jointplot(x=weather['humidity'], y=weather['dew'])
```

```python
sns.jointplot(x=weather['humidity'], y=weather['dew'], kind="hex")
```

```python
sns.jointplot(x=weather['humidity'], y=weather['dew'], kind="kde")
```

```python
sns.pairplot(weather[['humidity', 'dew', 'sealevelpressure']])
```

```python
sns.stripplot(x=weather['cloudcover'], y=weather['temp'])
```

```python
sns.stripplot(x=weather['cloudcover'], y=weather['temp'], jitter = True)
```

```python
sns.swarmplot(x=weather['humidity'], y=weather['temp'])
```

```python
sns.boxplot(x=weather['humidity'], y=weather['dew'], data=weather['temp'])
```

```python
sns.barplot(x=weather['humidity'], y=weather['dew'], data=weather['cloudcover'])
```

```python
sns.countplot(x=weather['cloudcover'])
```

```python
sns.pointplot(x=weather['humidity'], y=weather['temp'], data=weather['dew'])
```

```python
sns.lmplot(x="humidity", y="temp", data=weather)
```

```python
sns.lmplot(x="humidity", y="temp", hue="sealevelpressure",data=weather)
```
